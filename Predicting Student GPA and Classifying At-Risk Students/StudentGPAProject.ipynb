```python
def categorize(gpa):
    if gpa >= 3.5:
        return 0
    elif 3.0 <= gpa < 3.5:
        return 1
    elif 2.5 <= gpa < 3.0:
        return 2
    elif 2.0 <= gpa < 2.5:
        return 3
    else:
        return 4
gpadata['newGradeClass'] = gpadata['GPA'].apply(categorize)
```
